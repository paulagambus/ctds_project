## EDA

Notebook de carga y análisis inicial de los datos. Comparación de dataset grande y dataset pequeño

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#open dataset
main = pd.read_csv('data/movies.csv')
main
```

```python
#change 'name' column to 'title'
main.rename(columns={'name':'title'}, inplace=True)
```

```python
#open datasets for each platform
netflix = pd.read_csv('data/netflix_titles.csv')
amazon = pd.read_csv('data/amazon_prime_titles.csv')
hulu = pd.read_csv('data/hulu_titles.csv')
disney = pd.read_csv('data/disney_plus_titles.csv')
```

```python
netflix
```

```python
def remove_tv_show(df):
    return df[df['type'] == 'Movie']

print(netflix.shape)
print(amazon.shape)
print(hulu.shape)
print(disney.shape)

netflix = remove_tv_show(netflix)
disney = remove_tv_show(disney)
hulu = remove_tv_show(hulu)
amazon = remove_tv_show(amazon)

#dimension of the dataset
print('After removing shows')
print(netflix.shape)
print(amazon.shape)
print(hulu.shape)
print(disney.shape)
```

```python
#add platform column to main dataset
def platform_column(main, platform, platform_name):
    for index, row in main.iterrows():
        if row['title'] in platform['title'].values: #cambiar dependiendo del nombre de la columna
            main.loc[index, platform_name] = 1
        else:
            main.loc[index, platform_name] = 0
    return main

main_netflix = platform_column(main, netflix, 'Netflix')
main_net_ama = platform_column(main_netflix, amazon, 'Amazon')
main_net_ama_hulu = platform_column(main_net_ama, hulu, 'Hulu')
main_complete = platform_column(main_net_ama_hulu, disney, 'Disney')
```

```python
main_complete
```

```python
#count the number of movies in each platform
def count_movies(df, platform_name):
    return df[platform_name].sum()

netflix_count = count_movies(main_complete, 'Netflix')
print(netflix_count)
amazon_count = count_movies(main_complete, 'Amazon')
print(amazon_count)
hulu_count = count_movies(main_complete, 'Hulu')
print(hulu_count)
disney_count = count_movies(main_complete, 'Disney')
print(disney_count)
```

```python
#count the number of movies in more than one platform (take in consideration two, three and four platforms)
count = 0
for index, row in main_complete.iterrows():
    if row['Netflix'] + row['Amazon'] + row['Hulu'] + row['Disney'] > 1:
        count += 1
print(count)
```

```python
#identify which movies are in more than one platform
def more_than_one_platform(df):
    return df[(df['Netflix'] + df['Amazon'] + df['Hulu'] + df['Disney']) > 1]

in_platforms = more_than_one_platform(main_complete)
```

## Test with big dataset

```python
main_big = pd.read_csv('data/TMDB_all_movies.csv')
```

```python
#turn colum release_date into datetime
main_big['release_date'] = pd.to_datetime(main_big['release_date'])
main_big['year'] = main_big['release_date'].dt.year
main_big
```

```python
#keep movies released before 2021
main_big = main_big[main_big['year'] <= 2021]
```

```python
main_big
```

```python
#show rows with release_date = 2021
main_big[main_big['year'] == 2021] #checking if the filter worked
```

```python
bmain_netflix = platform_column(main_big, netflix, 'Netflix')
bmain_net_ama = platform_column(bmain_netflix, amazon, 'Amazon')
bmain_net_ama_hulu = platform_column(bmain_net_ama, hulu, 'Hulu')
bmain_complete = platform_column(bmain_net_ama_hulu, disney, 'Disney')
```

```python
#bmain_complete
```

```python
#count the number of movies in each platform
b_netflix_count = count_movies(bmain_complete, 'Netflix')
print(b_netflix_count)
b_amazon_count = count_movies(bmain_complete, 'Amazon')
print(b_amazon_count)
b_hulu_count = count_movies(bmain_complete, 'Hulu')
print(b_hulu_count)
b_disney_count = count_movies(bmain_complete, 'Disney')
print(b_disney_count)
```

```python
#plot comparing the number of movies in each platform in both datasets
platforms = ['Netflix', 'Amazon', 'Hulu', 'Disney']
movies = [netflix_count, amazon_count, hulu_count, disney_count]
b_movies = [b_netflix_count, b_amazon_count, b_hulu_count, b_disney_count]
#plot comparing the number of movies in each platform in both datasets
fig, ax = plt.subplots()
bar_width = 0.35
index = np.arange(len(platforms))
bar1 = plt.bar(index, movies, bar_width, label='Movies in small dataset')
bar2 = plt.bar(index + bar_width, b_movies, bar_width, label='Movies in big dataset')
plt.xlabel('Platforms')
plt.ylabel('Number of movies')
plt.title('Number of movies in each platform')
plt.xticks(index + bar_width/2, platforms)
plt.legend()
plt.show()
```
